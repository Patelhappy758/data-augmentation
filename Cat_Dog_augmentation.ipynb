```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img =image.load_img('/content/cat.jpg',target_size=(200,200))
```

```python
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='reflect'
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python

import os

i=0

# Create the directory if it doesn't exist
if not os.path.exists('aug'):
    os.makedirs('aug')

for output in datagen.flow(input_batch,batch_size=1,save_to_dir='aug'):

    i=i+1

    if i == 10:
      break
```

```python
input_batch.shape
```

```python
import os

# List the contents of the 'aug' directory
print(os.listdir('aug'))
```

```python
import matplotlib.pyplot as plt

# Get a list of all image filenames in the 'aug' directory
augmented_image_files = [os.path.join('aug', f) for f in os.listdir('aug') if f.endswith('.png')]

# Create a figure to display all images
plt.figure(figsize=(15, 10)) # Adjust figure size as needed

for i, img_path in enumerate(augmented_image_files):
    # Load the image
    img = image.load_img(img_path)

    # Create a subplot for each image
    plt.subplot(2, 5, i + 1) # Adjust subplot grid (e.g., 2 rows, 5 columns)
    plt.imshow(img)
    plt.title(f'Augmented Image {i+1}')
    plt.axis('off') # Hide axes for cleaner display

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```
